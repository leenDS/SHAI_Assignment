# **Task 4 : Basic Data Visualization**

```python
import matplotlib.pyplot as plt
plt.hist(df["TotalPay"],bins=30,color="lightblue",edgecolor="black")

plt.title("Total Pay (Salary)",fontsize=15)
plt.xlabel("Total Pay",color="blue")
plt.ylabel("Frequency",color="blue")
plt.show
```

```python
#The number of different Jobs present in the data
df['JobTitle'].nunique()
```

```python
#The most common jobs and how often each appears in the data
df['JobTitle'].value_counts().head(10)
```

```python
Job_counts = df['JobTitle'].value_counts()

Top_Job_Titles = Job_counts.nlargest(10)

threshold = 50

Other_Jobs = Job_counts[Job_counts < threshold].sum()

combined_categories = Top_Job_Titles.append(pd.Series({'Other': Other_Jobs}))

plt.figure(figsize=(11, 10))
plt.pie(combined_categories, labels=combined_categories.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Job Titles with "Other" Category')
plt.show()
```

# **Tasks 5 :Grouped Analysis**

```python
mean_BasePay= df_clean.groupby(['Year','JobTitle'])['BasePay'].mean().reset_index()
mean_BasePay
```

```python
#vizulation
top10_base=df_clean.groupby("JobTitle")["BasePay"].mean().sort_values(ascending=False).head(10).index
data_of_top10=df_clean[df_clean["JobTitle"].isin(top10_base)]
import seaborn as sns
plt.figure(figsize=(16,8))
sns.barplot(data=data_of_top10,x="JobTitle",y="BasePay",order=top10_base,width=0.5,hue="Year")
plt.xticks(rotation=30)
plt.tight_layout()
```

# **Task 6 : Simple Correlation Analysis**

```python
corr_= df_clean[["TotalPay"] + ["BasePay", "OvertimePay", "OtherPay","Year"]].corr()
print("Correlation between TotalPay and BasePay, OvertimePay, OtherPay, Year:")

plt.figure(figsize=(5, 4))
sns.heatmap(corr_,annot=True, cmap='pink', fmt=".2f",linewidths=0.9,linecolor='black')
plt.title('Correlation Heatmap')
plt.show()
```
